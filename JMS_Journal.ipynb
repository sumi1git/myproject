```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import StratifiedShuffleSplit, learning_curve, cross_val_score, KFold
from sklearn.preprocessing import LabelEncoder, StandardScaler, label_binarize
from sklearn.metrics import (
    confusion_matrix, classification_report, roc_auc_score, roc_curve
)

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.calibration import calibration_curve
from scipy.stats import f_oneway
from scipy.stats import ttest_rel, wilcoxon

import statsmodels.api as sm
```

```python
#df = pd.DataFrame(data)

# encode
le = LabelEncoder()
df["State_enc"] = le.fit_transform(df["State"])
classes = le.classes_

X = df[["Count"]]
y = df["State_enc"]


# ==========================================================
# 2. GUARANTEED BALANCED TEST SPLIT
# ==========================================================
sss = StratifiedShuffleSplit(n_splits=10, test_size=0.25, random_state=42)

for train_index, test_index in sss.split(X, y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

# scale
scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)


# ==========================================================
# 3. MODELS (NO LINEAR REGRESSION)
# ==========================================================
models = {
    "Logistic Regression": LogisticRegression(penalty='l2',
    C=1.0,
    #solver='lbfgs',
    max_iter=200
    #class_weight=None
),
    "SVM": SVC(C=.75, gamma='scale',probability=True
               kernel='rbf'
               ),
   # "Random Forest": RandomForestClassifier(n_estimators=200, random_state=42)
    "Random Forest": RandomForestClassifier(n_estimators = 200,
max_depth = 4,
min_samples_split = 4,
min_samples_leaf = 2,
max_features = "sqrt",
bootstrap = True)
}


# ==========================================================
# 4. UTILITY FUNCTIONS
# ==========================================================
def plot_confusion_matrix(cm, title):
    plt.figure(figsize=(5,4))
    sns.heatmap(cm, annot=True, cmap="Blues", fmt="d",
                xticklabels=classes, yticklabels=classes,annot_kws={"size": 16})
    plt.title(f"Confusion Matrix - {title}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

def plot_learning_curve(model, X, y, title):
    train_sizes, train_scores, val_scores = learning_curve(
        model, X, y, cv=8, scoring='accuracy',
        train_sizes=np.linspace(0.2, 1.0, 5)
    )
    plt.figure(figsize=(6,4))
    plt.plot(train_sizes, train_scores.mean(axis=1), label='Train')
    plt.plot(train_sizes, val_scores.mean(axis=1), label='Validation')
    plt.title(f"Learning Curve - {title}")
    plt.legend(); plt.grid(); plt.show()

def plot_calibration(model, X_test, y_test, title):
    y_prob = model.predict_proba(X_test)[:,2]  # for class GOOD
    prob_true, prob_pred = calibration_curve((y_test==2), y_prob, n_bins=5)
    plt.figure(figsize=(5,4))
    plt.plot(prob_pred, prob_true, marker='o')
    plt.plot([0,1],[0,1],'k--')
    plt.title(f"Calibration - {title}")
    plt.grid(); plt.show()

def plot_probability_distribution(ytr, yte, title):
    plt.figure(figsize=(6,4))
    sns.kdeplot(ytr, fill=True, label="Train")
    sns.kdeplot(yte, fill=True, label="Test")
    plt.title(f"Probability Distribution - {title}")
    plt.legend(); plt.show()


# ==========================================================
# 5. P-VALUE TEST (OLS REGRESSION)
# ==========================================================
print("\n==============================")
print("P-VALUE TEST (OLS)")
print("==============================")

X_sm = sm.add_constant(X)
ols = sm.OLS(y, X_sm).fit()
print(ols.summary())


# ==========================================================
# 6. MODEL TRAINING + GRAPHS + SIGNIFICANCE TESTS
# ==========================================================
all_scores = {}

for name, model in models.items():
    print("\n============================")
    print(f"MODEL: {name}")
    print("============================")

    model.fit(X_train_s, y_train)

    # Collect accuracy for ANOVA
    #kf = KFold(n_splits=2, shuffle=True, random_state=12)
    kf = KFold(n_splits=5, shuffle=True, random_state=12)
    #kf = KFold(n_splits=5, shuffle=True, random_state=12)
    scores = cross_val_score(model, X_train_s, y_train, cv=kf, scoring="accuracy")
    #scores = cross_val_score(model, X_train_s, y_train, scoring="accuracy")
    all_scores[name] = scores

    # Predictions
    y_pred = model.predict(X_test_s)
    y_prob = model.predict_proba(X_test_s)

    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    print("\nConfusion Matrix:\n", cm)
    plot_confusion_matrix(cm, name)

    # Classification report
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred, target_names=classes,digits=4))

    # ROC Curve
    plt.figure(figsize=(6,4))
    y_test_bin = label_binarize(y_test, classes=[0,1,2])
    for i, cls in enumerate(classes):
        fpr, tpr, _ = roc_curve(y_test_bin[:,i], y_prob[:,i])
        plt.plot(fpr, tpr, label=cls)
    plt.plot([0,1],[0,1],'k--')
    plt.title(f"ROC - {name}")
    plt.legend(); plt.grid(); plt.show()

    # Learning curve
    plot_learning_curve(model, X_train_s, y_train, name)

    # Calibration curve
    plot_calibration(model, X_test_s, y_test, name)

    # Probability distribution
    plot_probability_distribution(
        model.predict_proba(X_train_s)[:,2],
        model.predict_proba(X_test_s)[:,2],
        name
    )

    # Random Forest Feature Importance
    if name == "Random Forest":
        plt.figure(figsize=(4,3))
        plt.bar(["Count"], model.feature_importances_)
        plt.title("Feature Importance - RF")
        plt.show()

    # ==========================================================
    # 7. SIGNIFICANCE TESTS (T-TEST & WILCOXON)
    # ==========================================================
    print("\n------------------------------")
    print(f"Significance Tests for model: {name}")
    print("------------------------------")

    y_true = np.array(y_test)
    y_pred_arr = np.array(y_pred)

    # Paired t-test
    t_stat, p_val = ttest_rel(y_pred_arr, y_true)
    print("\nPaired T-test:")
    print("T-stat:", t_stat)
    print("p-value:", p_val)
    print("→ Significant difference" if p_val < 0.05 else "→ No significant difference")

    # Wilcoxon test
    try:
        z_stat, p_val_w = wilcoxon(y_pred_arr, y_true)
        print("\nWilcoxon Test:")
        print("Z-stat:", z_stat)
        print("p-value:", p_val_w)
        print("→ Significant difference" if p_val_w < 0.05 else "→ No significant difference")
    except ValueError:
        print("\nWilcoxon test could not run (likely identical samples).")


# ==========================================================
# 8. ANOVA — Compare model accuracies
# ==========================================================
print("\n==============================")
print("ANOVA TEST (MODEL ACCURACY)")
print("==============================")

anova = f_oneway(*all_scores.values())
print("F-statistic:", anova.statistic)
print("p-value:", anova.pvalue)

if anova.pvalue < 0.05:
    print("\n Models have significantly different performance.")
else:
    print("\nModels perform similarly (no significant difference).")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, learning_curve, cross_val_score, KFold
from sklearn.preprocessing import LabelEncoder, StandardScaler, label_binarize
from sklearn.metrics import (
    confusion_matrix, classification_report, roc_auc_score, roc_curve
)

from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.calibration import calibration_curve
from scipy.stats import f_oneway
import statsmodels.api as sm   # For p-values

# -------------------------------------------------------------------
# 1. LOAD DATA
# -------------------------------------------------------------------
data = {
    "Machine_Num": ["K0157-9","K0157-20","K0157-10","K0157-8","K0157-16","K0157-7","K0157-3",
                    "K0157-13","K0157-11","K0157-2","K0157-17","K0157-5","K0157-12","K0157-6",
                    "K0157-1","K0157-15","K0157-19","K0157-18","K0157-4","K0157-14"],
    "Count": [2184874,1985441,1789589,1784150,1746555,1668950,1645157,1588448,1486652,
              1251120,1221122,1112025,876520,805418,406595,365488,355489,102221,32954,1551],
    "State": ["BAD","BAD","FAIR","FAIR","FAIR","FAIR","GOOD","FAIR","GOOD","BAD",
              "GOOD","BAD","GOOD","GOOD","GOOD","GOOD","GOOD","GOOD","GOOD","GOOD"]
}
df = pd.DataFrame(data)

# Encode labels
le = LabelEncoder()
df["State_enc"] = le.fit_transform(df["State"])
classes = le.classes_

X = df[["Count"]]
y = df["State_enc"]

# -------------------------------------------------------------------
# 2. CUSTOM STRATIFIED SPLIT (2 samples per class in test set)
# -------------------------------------------------------------------
test_indices = []

for cls in df["State_enc"].unique():
    cls_indices = df[df["State_enc"] == cls].sample(2, random_state=42).index
    test_indices.extend(cls_indices)

test_df = df.loc[test_indices]
train_df = df.drop(test_indices)

print("\n==============================")
print("TEST SET (Exactly 2 from each class)")
print("==============================")
print(test_df)
print("\nClass distribution in test set:")
print(test_df["State"].value_counts())

X_train = train_df[["Count"]]
y_train = train_df["State_enc"]
X_test = test_df[["Count"]]
y_test = test_df["State_enc"]

# Scale
scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)

# -------------------------------------------------------------------
# MODELS
# -------------------------------------------------------------------
models = {
    "Logistic Regression": LogisticRegression(max_iter=2000),
    "SVM": SVC(probability=True),
    "Random Forest": RandomForestClassifier(n_estimators=200, random_state=42),
    "Linear Regression": LinearRegression()
}

# -------------------------------------------------------------------
# PLOTTING FUNCTIONS
# -------------------------------------------------------------------
def plot_confusion_matrix(cm, title):
    plt.figure(figsize=(5,4))
    sns.heatmap(
        cm, annot=True, cmap="Blues", fmt="d",
        xticklabels=classes, yticklabels=classes
    )
    plt.title(f"Confusion Matrix - {title}")
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.tight_layout()
    plt.show()

def plot_learning_curve(model, X, y, title):
    train_sizes, train_scores, val_scores = learning_curve(
        model, X, y,
        scoring='accuracy',
        cv=5,
        n_jobs=-1,
        train_sizes=np.linspace(0.2, 1.0, 5)
    )
    plt.figure(figsize=(6,4))
    plt.plot(train_sizes, train_scores.mean(axis=1), label="Train Accuracy")
    plt.plot(train_sizes, val_scores.mean(axis=1), label="Validation Accuracy")
    plt.title(f"Learning Curve - {title}")
    plt.xlabel("Training Samples")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()

def plot_probability_distribution(y_prob_train, y_prob_test, title):
    plt.figure(figsize=(6,4))
    sns.kdeplot(y_prob_train, label="Train Probability", fill=True)
    sns.kdeplot(y_prob_test, label="Test Probability", fill=True)
    plt.title(f"Predicted Probability Distribution - {title}")
    plt.xlabel("Probability of Class GOOD")
    plt.legend()
    plt.tight_layout()
    plt.show()

def plot_calibration(model, X_test, y_test, title):
    if not hasattr(model, "predict_proba"):
        return

    y_prob = model.predict_proba(X_test)[:,2]
    prob_true, prob_pred = calibration_curve((y_test==2), y_prob, n_bins=5)

    plt.figure(figsize=(5,4))
    plt.plot(prob_pred, prob_true, marker='o', label="Calibration")
    plt.plot([0,1],[0,1],'k--', label="Perfect Calibration")
    plt.title(f"Calibration Curve - {title}")
    plt.xlabel("Predicted Probability")
    plt.ylabel("True Probability")
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()

# -------------------------------------------------------------------
# 3. P-VALUE TEST (Statsmodels OLS)
# -------------------------------------------------------------------
print("\n==============================")
print("P-VALUE TEST (Feature Significance)")
print("==============================")

X_sm = sm.add_constant(X)
model_sm = sm.OLS(y, X_sm).fit()
print(model_sm.summary())


# -------------------------------------------------------------------
# 4. TRAIN MODELS
# -------------------------------------------------------------------
all_model_scores = {}

for name, model in models.items():
    print("\n============================")
    print(f"MODEL: {name}")
    print("============================")

    model.fit(X_train_s, y_train)

    # Collect cross-val scores
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    scores = cross_val_score(model, X_train_s, y_train, cv=kf, scoring="accuracy")
    all_model_scores[name] = scores

    # Linear Regression is NOT a classifier
    if name == "Linear Regression":
        print("Skipping classification metrics for Linear Regression.")
        continue

    # Predictions
    y_pred = model.predict(X_test_s)
    y_prob = model.predict_proba(X_test_s)

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    print("\nConfusion Matrix:")
    print(cm)
    plot_confusion_matrix(cm, name)

    # Classification Report
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred, target_names=classes))

    # AUC (Multiclass)
    y_test_bin = label_binarize(y_test, classes=[0,1,2])
    auc = roc_auc_score(y_test_bin, y_prob, multi_class="ovr")
    print(f"AUC Score: {auc:.4f}")

    # ROC Curve
    plt.figure(figsize=(6,4))
    for i, cls in enumerate(classes):
        fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_prob[:, i])
        plt.plot(fpr, tpr, label=f"{cls}")
    plt.plot([0,1],[0,1],'k--')
    plt.title(f"ROC Curve - {name}")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()

    # Learning Curve
    plot_learning_curve(model, X_train_s, y_train, name)

    # Calibration Curve
    plot_calibration(model, X_test_s, y_test, name)

    # Probability Distribution for GOOD class
    y_prob_train = model.predict_proba(X_train_s)[:,2]
    y_prob_test = model.predict_proba(X_test_s)[:,2]
    plot_probability_distribution(y_prob_train, y_prob_test, name)

    # Feature Importance (RF only)
    if name == "Random Forest":
        plt.figure(figsize=(4,3))
        plt.bar(["Count"], model.feature_importances_)
        plt.title("Feature Importance - Random Forest")
        plt.tight_layout()
        plt.show()

# -------------------------------------------------------------------
# 5. ANOVA TEST ACROSS MODELS
# -------------------------------------------------------------------
print("\n==============================")
print("ANOVA TEST ON MODEL ACCURACY")
print("==============================")

anova = f_oneway(*all_model_scores.values())
print("\nANOVA F-statistic:", anova.statistic)
print("ANOVA p-value:", anova.pvalue)

if anova.pvalue < 0.05:
    print("\n🔥 At least one model significantly outperforms others.")
else:
    print("\nModels perform similarly (no significant difference).")

```
